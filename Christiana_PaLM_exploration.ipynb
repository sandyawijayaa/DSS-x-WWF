```python
def embed_fn(text):
  return palm.generate_embeddings(model=embed_model, text=text)['embedding']
```

```python
def chunk_text(text, byte_limit=100000):
    char_limit = byte_limit // max(len(s.encode('utf-8')) for s in text)
    return [text[i:i+char_limit] for i in range(0, len(text), char_limit)]

def get_batch_texts(texts, byte_limit=100000):
    batch = []
    current_batch_byte_size = 0
    for text in texts:
        if len(text.encode('utf-8')) > byte_limit:
            for chunk in chunk_text(text, byte_limit):
                yield [chunk]
        else:
            if current_batch_byte_size + len(text.encode('utf-8')) > byte_limit:
                if batch:
                    yield batch
                batch = []
                current_batch_byte_size = 0
            batch.append(text)
            current_batch_byte_size += len(text.encode('utf-8'))
    if batch:
        yield batch

embeddings = []
for batch_texts in get_batch_texts(df_palm['Text']):
    batch_embeddings = embed_fn(batch_texts)
    if not batch_embeddings:
        print("No embeddings returned for texts:", batch_texts)
        batch_embeddings = [None] * len(batch_texts)  # Placeholder

    if len(batch_texts) == 1:
        original_text_length = len(df_palm['Text'].iloc[len(embeddings)])
        if len(batch_texts[0]) != original_text_length:
            # If this is a chunk of a larger text, concatenate embeddings
            if embeddings:  # Check if embeddings isn't empty to prevent index error
                embeddings[-1] += batch_embeddings[0]
            else:
                print("Error: Embeddings list is empty!")
                embeddings.append(batch_embeddings[0])
        else:
            embeddings.extend(batch_embeddings)
    else:
        embeddings.extend(batch_embeddings)

df_palm['Embeddings'] = embeddings
```

```python
def find_most_relevant_segment(query, text):
    """
    Given a query and a text, find the most relevant segment of the text in relation to the query.
    """
    # Split the text into segments, for simplicity, let's consider segments as sentences.
    segments = text.split('.')  # Splitting by periods to get sentences

    # Compute embeddings for each segment
    segment_embeddings = [palm.generate_embeddings(model=embed_model, text=segment)['embedding'] for segment in segments]

    # Compute embedding for the query
    query_embedding = palm.generate_embeddings(model=embed_model, text=query)['embedding']

    # Find the segment whose embedding has the highest dot product with the query embedding
    dot_products = np.dot(segment_embeddings, query_embedding)
    idx_max = np.argmax(dot_products)

    return segments[idx_max]
```

```python
def make_prompt(query, relevant_passage):
  escaped = relevant_passage.replace("'", "").replace('"', "").replace("\n", " ")
  prompt = textwrap.dedent("""You are a helpful and informative bot that answers questions using text from the reference passage included below. \
  Be sure to not make up fake information if you can not find relavent information in the text. \
  QUESTION: '{query}'
  PASSAGE: '{relevant_passage}'

    ANSWER:
  """).format(query=query, relevant_passage=escaped)

  return prompt
```

```python
text_models = [m for m in palm.list_models() if 'generateText' in m.supported_generation_methods]

text_model = text_models[0]
temperature = 0.5
```

```python
def get_answer_for_query_using_prompt(query, relevant_segment, text_model):
    """
    Generate an answer using the prompt-based approach.
    """
    prompt = make_prompt(query, relevant_segment)
    answer = palm.generate_text(prompt=prompt, model=text_model, candidate_count=1, temperature=0.7, max_output_tokens=1000)
    return answer.candidates[0]['output'].strip()
```

```python
queries = [
    "What location or locations did this project take place in?",
    "What was the impact of the work in this region?",
    "What are potential next steps for this project?"
]
```

```python
results = []
for pdf_name in samplefNames:
    answers = [pdf_name]  # Start with the PDF name

    # Load the specific PDF
    pdf_loader = UnstructuredPDFLoader(os.path.join(pdf_folder_path, pdf_name))
    pdf_index = VectorstoreIndexCreator(
        embedding=GooglePalmEmbeddings(),
        text_splitter=CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
    ).from_loaders([pdf_loader])

    # For each query, we will re-instantiate the `pdf_chain` object
    for query in queries:
        pdf_chain = RetrievalQA.from_chain_type(llm=llm,
                                chain_type="stuff",
                                retriever=pdf_index.vectorstore.as_retriever(),
                                input_key="question")
        answer = pdf_chain.run(query)
        answers.append(answer)  # Append the answer to the list

    # Append the answers to the results list
    results.append(answers)
```
