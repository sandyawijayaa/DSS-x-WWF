given folderpath and filenames, return dataframe with columns name(name of pdf file) and text (full text of pdf)

```python
def extractText(folderpath, filenames):
    dInformation = {"Name": [], "Text": []}

    for fName in filenames:
        # Construct full file path
        full_path = os.path.join(folderpath, fName)

        # Check if the file exists and is a PDF
        if os.path.exists(full_path) and fName.lower().endswith('.pdf'):
            fullText = ""
            try:
                with pdfplumber.open(full_path) as pdf:
                    for page in pdf.pages:
                        text = page.extract_text()
                        if text:
                            text = text.replace("\n", " ")
                            fullText += (text + " ")

                dInformation["Name"].append(fName)
                dInformation["Text"].append(fullText)

            except Exception as e:
                print(f"An error occurred with file {fName}: {e}")
                dInformation["Name"].append(fName)
                dInformation["Text"].append("")  # Append empty string if there was an error
        else:
            print(f"File not found or is not a PDF: {fName}")

    df = pd.DataFrame(dInformation)
    return df
```

Note: If lengths of texts used as the examples in the prompt are greater than ~15000, max tokens are exceeded

```python
text_models = [m for m in palm.list_models() if 'generateText' in m.supported_generation_methods]
text_model = text_models[0]
temperature=0.2
```

Run this cell to use the prompt defined from the make_prompt cell to query on each text in df.

```python
answers_accomplishment=[]
for index, row in df.iterrows():
    text = row["Text"]
    chunks = [text[i:i + 32000] for i in range(0, len(text), 32000) if text[i:i + 32000].strip()]

    answer_text = ''
    for chunk in chunks:
        if not answer_text:
            prompt = make_prompt("Identify the key accomplishments/performance metrics/results of this pdf. The results should pertrain to the work of World wildlife fund(WWF) in 1 or multiple regions. Limit your response to 3 sentences and make it concise.", chunk, df)
            answer = palm.generate_text(prompt=prompt,
                                        model=text_model,
                                        candidate_count=1,
                                        temperature=temperature,
                                        max_output_tokens=1000)

            if answer.result:
                answer_text = answer.result
                answers_accomplishment.append(answer_text)
                break

    if not answer_text:
        answers_accomplishment.append("No relevant information found")
```
