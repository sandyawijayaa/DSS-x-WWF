```python
import pprint
import google.generativeai as palm
import pandas as pd
from PyPDF2 import PdfReader
import numpy as np
import textwrap
import pdf2image
import os
import pdfplumber
```

```python
def extractText(folderpath, filenames):
    dInformation = {"Name": [], "Text": []}

    for fName in filenames:
        # Construct full file path
        full_path = os.path.join(folderpath, fName)

        # Check if the file exists and is a PDF
        if os.path.exists(full_path) and fName.lower().endswith('.pdf'):
            fullText = ""
            try:
                with pdfplumber.open(full_path) as pdf:
                    for page in pdf.pages:
                        text = page.extract_text()
                        if text:
                            text = text.replace("\n", " ")
                            fullText += (text + " ")

                dInformation["Name"].append(fName)
                dInformation["Text"].append(fullText)

            except Exception as e:
                print(f"An error occurred with file {fName}: {e}")
                dInformation["Name"].append(fName)
                dInformation["Text"].append("")  # Append empty string if there was an error
        else:
            print(f"File not found or is not a PDF: {fName}")

    df = pd.DataFrame(dInformation)
    return df
```

```python
text_models = [m for m in palm.list_models() if 'generateText' in m.supported_generation_methods]
text_model = text_models[0]
temperature=0.2
```

input token limit:8k. For the PaLM 2 model, token is equivalent to about 4 characters.

```python
answers_locations=[]
for index, row in df.iterrows():
    text = row["Text"]
    chunks = [text[i:i + 32000] for i in range(0, len(text), 32000) if text[i:i + 32000].strip()]

    answer_text = ''
    for chunk in chunks:
        if not answer_text:
            prompt = make_prompt("Tell me the location of the project", chunk)
            answer = palm.generate_text(prompt=prompt,
                                        model=text_model,
                                        candidate_count=1,
                                        temperature=temperature,
                                        max_output_tokens=1000)

            # Check if answer is not empty
            if answer.result:
                answer_text = answer.result
                answers_locations.append(answer_text)
                break

    if not answer_text:
        answers_locations.append("No relevant information found")
```
