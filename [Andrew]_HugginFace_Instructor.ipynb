```python
import os
import re

import numpy as np
import pandas as pd
from PyPDF2 import PdfReader
```

```python
from zipfile import ZipFile

with ZipFile("/content/WWFxDSS/pdfs_sample.zip", 'r') as zObject:
    zObject.extractall(
        path="/content/WWFxDSS/pdfs_sample")
```

```python
fnames = []
folderpath = "/content/WWFxDSS"
for fname in os.listdir(folderpath):
  fnames.append(fname)

def extractText(fNames):
  dInformation = {"Name": [], "Text": []}
  for fName in fNames:
    if not fName.endswith(".pdf"):
      continue
    fPath = os.path.join(folderpath, fName)
    reader = PdfReader(fPath)
    fullText = ""
    for pgn in range(len(reader.pages)):
      page = reader.pages[pgn]
      text = page.extract_text()
      text = text.replace("\n", "")
      fullText += (text + " ")
    fName = re.sub(r'\W+', '', fName[:-4])
    dInformation["Name"].append(fName)
    dInformation["Text"].append(fullText)
  df = pd.DataFrame(dInformation)
  return df

df = extractText(fnames)
display(df)
```

```python
def create_vectordbs(embedder, splitter):
    all_vectordbs = {}

    for i, row in df.iterrows():
        texts = splitter.create_documents([row["Text"]])
        splits = splitter.split_documents(texts)

        all_vectordbs[row["Name"]] = Chroma.from_documents(
            documents=splits,
            embedding=embedder,
            collection_name=row["Name"]
        )

    return all_vectordbs
```

```python
instruct_vector_dbs = create_vectordbs(instructor_embedder, r_splitter)
```

```python
def answer_questions(vector_dbs, question):
  answer_docs = {}
  for name, vector_db in vector_dbs.items():
    fetch_k = len(vector_db.get()['documents'])
    response = vector_db.max_marginal_relevance_search(question, fetch_k = fetch_k, k = 1)
    answer_docs[name] = response[0].page_content
  return answer_docs
```

```python
question = "What was the location our project was conducted in?"

location_docs = answer_questions(instruct_vector_dbs, question)
```

```python
question = "What are the primary accomplishments of our project?"

accomplishment_docs = answer_questions(instruct_vector_dbs, question)
```

```python
df['Location_docs'] = location_docs.values()
df['Accomplishment_docs'] = accomplishment_docs.values()
```

```python
import spacy
nlp = spacy.load('en_core_web_lg')
```

```python
def getLocations(df, docs):
  df["Location"] = ""
  for name, doc in docs.items():
    doc = nlp(doc)
    for ent in doc.ents:
      if (ent.label_ == 'GPE' or ent.label_ == 'LOC'):
        df.loc[df["Name"] == name, "Location"] += ent.text
  return df
```

```python
getLocations(df, location_docs)
```

```python
def updateAccomplishments(df, docs):
  df["Accomplishments"] = ""
  for name, doc in docs.items():
    sentences = doc.split(".")
    for sentence in sentences:
      tokens = nlp(sentence)
      for ent in tokens.ents:
        if (ent.label_ in ['DATE', 'EVENT', 'LAW', 'MONEY', 'ORG', 'PERCENT', 'QUANTITY']):
          df.loc[df["Name"] == name, "Accomplishments"] += sentence
  return df
```

```python
def get_llama_response(prompt: str):
    sequences = llama_pipeline(
        prompt,
        do_sample=True,
        top_k=10,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_length=512,
    )
    return sequences[0]['generated_text'].partition("ANSWer: ")[-1]
```
