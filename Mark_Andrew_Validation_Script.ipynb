```python
from typing import TextIO
keyphrases = kw_model.extract_keywords(posTest, keyphrase_ngram_range=(1, 1), top_n=10)
print(keyphrases)
```

```python
def calculateScore(outputSeries, testSeries):
  totalScore = 0
  for i in range(min(len(outputSeries), len(testSeries))):
    if outputSeries[i] != outputSeries[i] or testSeries[i] != testSeries[i]:
      print("NaN Value ==>", end=" ")
      continue
    try:
      output_embeddings, output_word_embeddings = kw_model.extract_embeddings(outputSeries[i], min_df=1, stop_words="english")
      test_embeddings, test_word_embeddings = kw_model.extract_embeddings(testSeries[i], min_df=1, stop_words="english")
      score = metrics.pairwise.cosine_similarity(output_word_embeddings, test_word_embeddings)
      print(str(score.mean()) + " ==>", end=" ")
      totalScore += score.mean()
    except:
      print("Misc Error ==>", end=" ")
      continue
  print()
  return totalScore
```
